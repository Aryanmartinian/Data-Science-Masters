```python
# python function - palindrome - 121 - 121
def is_palindrome(s):
    s = str(s)
    if s==s[::-1]:
        return True
    else:
        return False
    
print(is_palindrome(121))
print(is_palindrome(231))
```

```python
# Occurrences of each number in a list - 

def count_occurrences(numbers):
    freq = {} 
    for num in numbers:
        if num in freq:
            freq[num]+=1
        else:
            freq[num]=1
    
    for key,value in freq.items():
        print(f'{key} occurs {value} times')


numbers = [1,2,3,1,4,4,5,6,7,7]
count_occurrences(numbers)
```
